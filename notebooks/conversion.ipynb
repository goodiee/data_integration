```python
import pandas as pd
import os

input_file = "M:/accenture/final_project/covid19-platform/data/gdp_long.csv"
output_file = "data/gdp_ppp_long.csv"

gdp = pd.read_csv(input_file)
gdp.columns = gdp.columns.str.strip() 

gdp_long = gdp.melt(
    id_vars=["Sr.No", "Country"],
    var_name="Year",
    value_name="GDP_PPP_Per_Capita"
)

gdp_long = gdp_long.dropna(subset=["GDP_PPP_Per_Capita"])

gdp_long["Year"] = gdp_long["Year"].astype(int)

os.makedirs("data", exist_ok=True)
gdp_long.to_csv(output_file, index=False)

print(f"Saved long GDP data to {output_file}")
print(gdp_long.head())
```
